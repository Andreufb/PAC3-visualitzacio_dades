```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get the CSV data
data = pd.read_csv("owid-co2-data.csv") 


# Function to clean the data
def clean_data_country_evolution(value):
    # Lists and dictionatry to save the data
    countries = []
    region = []
    image = []
    years = {}

    # iteration over data to get it to the correct form
    for row in data.index:
        if data['country'][row] not in countries:
            countries.append(data['country'][row])
            region.append('')
            image.append('')
        if data['year'][row] not in years:
             years[data['year'][row]] = {}
        years[data['year'][row]].update({data['country'][row]: data[value][row]})
    
    total_data = {
        'countries': countries,
        'region': region,
        'image': image
    }
    
    # Iteration over data to add to the JSON in correct structure
    for year, values in years.items():
        total_data[year] = []
        for country in total_data['countries']:
            found = False
            for year_country, value in values.items():
                if country == year_country:
                    found = True
                    total_data[year].append(value)
            if not found:
                total_data[year].append(None)
    
    return total_data

```

```python
clean_data = clean_data_country_evolution('population')
```

```python
import base64
from IPython.display import HTML

def create_download_link( df, title = "Download CSV file", filename = "clean_data.csv"):
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
df = pd.DataFrame(clean_data, columns=list(clean_data.keys()))
create_download_link(df)
```
